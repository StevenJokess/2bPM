# 认知智能

2020年，是人工智能领域的扎实落地之年，也是从感知智能迈向认知智能的规模突破之年。知识图谱技术与图计算应用场景取得爆发性增长，从增强自然语言能力、人工智能模型可解释性、机器学习能力提升三个维度助力各大企业机构向认知智能领域演变。[1]

如果说感知智能时代，人工智能三要素是数据，算法，算力。

那么在认知智能时代，三要素则需要进一步迭代与提升，具体来说就是：

- 算法不是关键，工程架构才是关键；
- 算力不是关键，人机协作才是关键；
- 数据(量)不是关键，数据的结构才是关键；
